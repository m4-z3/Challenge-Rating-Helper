# Data Masters: DnD Challenge Rating Helper

### What is DnD?
Dungeons and Dragons (DnD) is a fantasy tabletop role-playing game. Players called Dungeon Masters (DMs) lead groups, or parties, of other players through adventures involving various activities, including fighting monsters.

### Explaining what's in a stat block
**Stat Block**- A stat block is a chart describing of all of a monster’s stats. It contains all the information you would need to know about what the monster is capable of in the game.

<img src="stat_block_example.png">

**1) Name**- Not relevant to our analysis. Just the name of the monster, nothing special.

**2) Size**- The possible size values, in ascending order, are: Tiny, Small, Medium, Large, Huge, Gargantuan. It primarily determines the amount of space a monster occupies. It also has some impact on how easy the monster is to hit and how much damage the monster can do, with bigger monsters being easier to hit and doing more damage.

**3) Type, Subtype**- Types and subtypes can be thought of as families of species. Monsters of the same type often share similar characteristics. For example, dragon type monsters are typically big, powerful, intelligent, and highly magical.

**4) Alignment**- Not relevant to our analysis. Alignment is essentially the monster’s moral compass. Are they good or evil? Do they stick to the rules (lawful) or do what feels right to them (chaotic)? Or are they neutral, somewhere in between?

**5) Armor Class (AC)**- AC is an integer determining how hard it is to hit a monster. Higher numbers are harder to hit.

**6) Health Points (HP)**- HP is an integer representing how much damage a monster can take before it dies. There is also a value in parenthesis representing the dice calculation for the HP, but we don’t need to worry about this since we’re focusing on the average health.

**7) Speed**- Speed is how far a monster can move in a single round (with each round being approximately 6 seconds), measured in feet. Most monsters only have a walking speed, but some can also burrow, climb, fly, or swim, and may have a different speed when moving in those ways.

**8) Ability Scores**- This is a set of six integers representing strength (STR), dexterity (DEX), constitution (CON), intelligence (INT), wisdom (WIS), and charisma (CHA). These scores affect many aspects of a monster, such as how hard it can hit, how well it can avoid an attack, how powerful are its spells and actions, and how well it can avoid or resist spells, actions, and other conditions. Each ability score also has an additional value in parentheses, representing the bonus that ability score gives to saving throws and skill checks. Since the bonus is simply a function of the ability score, we don’t need to worry about the bonus.

**9) Saving Throw Proficiencies**- Saving throws occur when a monster needs to avoid some effect, such as from a spell. Saving throws check one of the six ability scores, where the monster needs to score higher than a specified number to avoid the effect. Being proficient in a saving throw allows the monster to have an additional bonus added to their score, making it easier to avoid the effect. Not all monsters have saving throw proficiencies.

**10) Skill Proficiencies**- Skill checks function similarly to saving throws, but it occurs when a monster is trying to complete an action. Skill checks check one of the skills, which are categorized under and use the same score as the ability scores. The monster needs to score higher than a specified number to successfully complete the action. Being proficient in a skill allows the monster to have an additional bonus added to their score, making it easier to complete the action. Not all monsters have skill proficiencies.

List of skills and their ability score: Acrobatics (DEX), Animal Handling (WIS), Arcana (INT), Athletics (STR), Deception (CHA), History (INT), Insight (WIS), Intimidation (CHA), Investigation (INT), Medicine (WIS), Nature (INT), Perception (WIS), Performance (CHA), Persuasion (CHA), Religion (INT), Sleight of Hand (DEX), Stealth (DEX), Survival (WIS)

**11) Damage Resistance, Vulnerability, Immunity**- There are multiple different types of damage. Being resistant to a type of damage makes the monster take only half the damage from that type. Being vulnerable to a type of damage makes the monster take double the damage from that type. Being immune to a type of damage makes the monster take no damage from that type. Not all monsters have damage resistances, vulnerabilities, or immunities.

List of damage types: acid, bludgeoning, cold, fire, force, lightning, necrotic, piercing, poison, psychic, radiant, slashing, thunder

**12) Condition Immunity**- There are multiple different types of conditions. Conditions disable a monster in some form, such as preventing movement or actions. Being immune to a type of condition makes the monster never get that condition.

List of some common condition types: blinded, charmed, deafened, grappled, incapacitated, paralyzed, petrified, poisoned, prone, restrained, stunned, unconscious, exhaustion

**13) Senses**- Senses specifies special senses a monster might have that will help it perceive its surroundings. All monsters have a passive perception, based on their perception (WIS) skill. Some monsters might have blindsight, darkvision, tremorsense, or truesight, which typically have a range of effectiveness. Blindsight refers to any method of perceiving surroundings without sight, such as echolocation. Darkvision refers to the ability to seek in dark and dim light. Tremorsense refers to the ability to sense surroundings through vibrations within a surface, differing from blindsight in that it can't sense in the air, but can sense through surfaces. Truesight refers to the ability to see though invisibility and illusions.

**14) Languages**- Not relevant to our analysis. This includes all languages the monster speaks and understands, as well as if it is capable of telepathy.

**15) Challenge Rating (CR)**- CR represents how difficult a monster is to defeat. CR is an integer from 0 to 30 inclusive or the values 1/8, 1/4, or 1/2, with 0 representing trivial monsters and 30 representing the most powerful monsters. In general, a monster at CR X is a good match for the average party of 4 players at level X.

**16) Experience Points (XP)**- Not relevant to our analysis. When players defeat monsters, they get an amount of XP. Monsters with higher CR give more XP.

**17) Legendary Resistance**- Some very powerful monsters (thus the name "legendary") have legendary resistance. Legendary resistance lets a monster automatically pass a saving throw, which is very useful for avoiding conditions or damage. A monster with legendary resistance will have a limited number of uses per day. Not all monsters have Legendary resistance.

**18) Actions**- Actions represent special moves that a monster can do. We split this into attack type actions and other actions. Actions are very varied, so outside of attacks, there isn’t much that can be done to categorize them. Attacks generally have a chance to miss, are only effective within a certain range, and do damage. Other actions can do many things, from performing multiple attacks at once, to modifying the environment, or applying conditions or effects to enemies.

**19) Reactions**- Reactions are a specialized action that a monster can do as a response when it is attacked. Like actions, they have varied effects. Not all monsters have reactions.

**20) Legendary Actions**- Some very powerful monsters (thus the name "legendary") have legendary actions. Legendary actions are a specialized action that a monster can do when it is another creature’s turn. Like actions, they have varied effects. A monster with legendary actions will have a limited number of uses per round. Not all monsters have Legendary actions.

### Our Question
Given our own experience playing DnD, we sometimes encounter moments when the DM has set up an encounter that was too hard and the whole party of players ends up knocked out. This is not fun and could lead to discontent towards the DM for misjudging how strong a monster is. So our questions is, how do we help dungeon masters create appropriately challenged encounters? Conveniently, DnD has a stat called challenge rating which is perfect for measuring how strong a monster will be. As such, we decided to shift our question: how exactly is the challenge rating related to the various stat block metrics?

### What are challenge ratings?
The difficulty of monsters is described by a challenge rating (CR). CR is an integer from 0 to 30 inclusive or the values 1/8, 1/4, or 1/2, with 0 representing trivial monsters and 30 representing the most powerful monsters. In general, a monster at CR X is a good match for the average party of 4 players at level X.

Note that the power of players, as measured by their level, is not continuous. Players gain levels by doing various activities to collect experience points and increase their level at specific experience point milestones. Until they reach that milestone, there is no benefit from collecting experience points. Once a player increases their level, they receive new abilities and increase their health, abilities, and other useful features. This results in discontinuous jumps in power between levels. CR is related to player level, so it also has these discontinuous jumps in power.


### Our Goal
We want to be able to look at various aspects of a monster's stats and create a model that can predict the CR of a monster when given its stats. Ideally, this model could be applied to check the CR of any monster that a DM creates and wants to use.

Given DnD’s history, there’s already a lot of resources online for DMs to use in order to figure out how to set up appropriate encounters and figuring out the challenge rating (such as http://kobold.club/fight/#/encounter-builder, https://5e.tools/crcalculator.html, and the Dungeon Master’s Guide). However, these resources only give a general guideline based on existing monsters or focusing on a select few features in determining challenge rating (such as hit points, armor class, and damage). ## How did we obtain the data?

The initial obvious choice for scraping our data would be the official Monster Manual. However, this source alone would be limited only several hundreds of monsters, rather than the thousands needed for a good analysis. It was also quite difficult to deal with the optical character recognition for the pdf of the book.

Instead, we chose to turn to fan wikis and databases, which included some official monsters, but also many unofficial, also known as homebrewed, monsters. This will also make our data more reflective of real games, since many players like to homebrew their own monsters to add into their games.
We started with a cursory skim through some of the more popular and larger wikis and databases. We wanted to find one that had a lot of data, but was also formatted in a way that was easy to scrape from.

The site we chose was https://www.dandwiki.com/wiki/. It has a few official entries and many homebrew entries. All of our scraping was done with BeautifulSoup and regex, and our data was stored in a .tsv file.

Since this is a wiki with entries from many different authors, multiple variations and errors in formatting should be expected. We started by only scraping data that matched the formatting of the template entries, and then gradually expanded to include various variations of the formatting to get as much of the data as possible. Any entries that were missing important values or otherwise too different from the proper formatting were discarded.

Here’s a few examples of some of the bigger formatting issues (some of which were recoverable, some of which were not):

We also had to filter out various troll or joke entries. Fortunately for us, the wiki has a special “April Fools!” tag to designate such entries. Since the author has to add this tag themselves, it is possible that some joke entries don’t have this tag. We manually removed any entries that seemed extremely unusual or strangely formatted. We also chose to filter out entries with a CR that was larger than the official cap of 30, or any other value that was out of the standard range of values. For entries with a CR greater than 30, a lot of them were overpowered monsters that might be imbalanced. Other values that were out of standard ranges indicated the monster could be imbalanced or would be difficult to categorize among the rest of the data for analysis.

## Logic behind preprocessing

After looking over our data, we decided to manually eliminate some features after determining that they were insignificant/ redundant in contributing to the challenge rating. These features were categorized by the fact that they contributed to other features, were very situational and didn’t apply to a majority of the monsters, or were dependent on the creativity of the dungeon master (which we chose not to include as our goal isn’t to control how a dungeon master will create their monster). We chose to then eliminate the following features:

- Features that contributed to another feature
    - hit die num
    - hit die type
    - hit die bonus
    - XP (relates to challenge rating)

- Situational features
    - alternate AC (only applicable to shape shifters)
    - alternate  speed
    - alternate burrow speed
    - alternate climb speed
    - alternate flying speed
    - alternate swim speed
    - boss actions (only appears in homebrew and not applicable to the official monsters)

- Features dependant on dungeon master
    - language
    - alignment
    - armor type
    - type
    - sub type

We then converted the ordinal features and categorical features to numerical values. This meant numbering the ordinal features and creating separate columns for boolean categorical features. The following shows the features in which this applies to:

- Ordinal features
    - size
- Categorical features
    - saving throw proficiencies
    - skill proficiencies
    - type of movement

Additionally some features contained more information in them that needed to be processed into separate columns to get a better numerical picture.

- Features divided into separate columns
    - senses
    - legendary actions
    - actions

On top of that, we deemed some of these features which contained more information would be best as generalized features that only counted the number of values it contained. In this way we can reduce any bias towards certain types of values that should really be up to the dungeon master’s discretion.

- Features that translate to counted values
    - damage vulnerability
    - damage resistance,
    - damage immunity
    - condition immunity
    - action
    - reaction

Finally for speed we decided that we would look at the summed value given to speed and use boolean values to denote which type of speed contributed to the total (whether it’s walking, burrowing, climbing, flying, or swimming). We did this in the hopes of generalizing speed and not wanting the type of speed to contribute as heavily (since once again, this should mostly be up to the dungeon master who created the monster).

```
import string
import re
import ast

df = pd.read_csv('monster_data.tsv', sep='\t')

# dropped features
remove_list = ['hit die num','hit die type','hit die bonus','language','alignment','armor type','type','subtype','XP', 'boss actions', 'alternate AC', 'alternate speed', 'alternate burrow speed', 'alternate climb speed', 'alternate flying speed', 'alternate swim speed']
df = df.drop(remove_list, axis = 1)

# monsters that need to be removed due to unreliable data values (ex. monsters that were created but abandoned by their creator)
removed_monsters = ['Azorius']
df = df[~df['name'].isin(removed_monsters)]

# monsters were removed due to incorrect scraping (making their values unreliable)
df = df[df['hit point avg'] != 'none']

# change size to ordinal numbers using official size categories: tiny, small, medium, large, huge, and gargantuan
# monsters with sizes that don't fit these will be discarded
#      - this is because these most likely homebrew content that didn't follow the conventional dnd format, making them unreliable in terms of guiding inexperienced DMs
size_map = {
    'tiny': 1,
    'small': 2,
    'medium': 3,
    'large': 4,
    'huge': 5,
    'gargantuan': 6
}

df['size'] = df['size'].map(lambda x: size_map[x.lower()] if x.lower() in size_map else -1)
df = df[df['size'] != -1]

# separating saving throw proficiencies into separate columns
def find_proficiencies(pattern):
  def record(x):
    if x == 'none':
      return 0
    if re.search(pattern, x.lower()):
      return 1
    return 0
  return record

df['str saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('str[a-z]*'))
df['dex saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('dex[a-z]*'))
df['con saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('con[a-z]*'))
df['int saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('int[a-z]*'))
df['wis saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('wis[a-z]*'))
df['cha saving throw'] = df['saving throws proficiencies'].map(find_proficiencies('cha[a-z]*'))

# separate skill proficiencies into different columns
df['athletics'] = df['skill proficiencies'].map(find_proficiencies('athletics +'))
df['acrobatics'] = df['skill proficiencies'].map(find_proficiencies('acrobatics +'))
df['sleight of hand'] = df['skill proficiencies'].map(find_proficiencies('sleight +of +hand +'))
df['stealth'] = df['skill proficiencies'].map(find_proficiencies('stealth +'))
df['arcana'] = df['skill proficiencies'].map(find_proficiencies('arcana +'))
df['history'] = df['skill proficiencies'].map(find_proficiencies('history +'))
df['investigation'] = df['skill proficiencies'].map(find_proficiencies('investigation +'))
df['nature'] = df['skill proficiencies'].map(find_proficiencies('nature +'))
df['religion'] = df['skill proficiencies'].map(find_proficiencies('religion +'))
df['animal handling'] = df['skill proficiencies'].map(find_proficiencies('animal +handling +'))
df['insight'] = df['skill proficiencies'].map(find_proficiencies('insight +'))
df['medicine'] = df['skill proficiencies'].map(find_proficiencies('medicine +'))
df['perception'] = df['skill proficiencies'].map(find_proficiencies('perception +'))
df['survival'] = df['skill proficiencies'].map(find_proficiencies('survival +'))
df['deception'] = df['skill proficiencies'].map(find_proficiencies('deception +'))
df['intimidation'] = df['skill proficiencies'].map(find_proficiencies('intimidation +'))
df['performance'] = df['skill proficiencies'].map(find_proficiencies('performance +'))
df['persuasion'] = df['skill proficiencies'].map(find_proficiencies('persuasion +'))

# separate the senses into different columns
def find(pattern):
  def record(x):
    matched = pattern.search(x.lower())
    if matched:
      return matched.group(1)
    return 0
  return record

pass_perception_pattern = re.compile(r'passive +perception +([0-9]+)')
blindsight_pattern = re.compile(r'blindsight +([0-9]+)')
darkvision_pattern = re.compile(r'darkvision +([0-9]+)')
tremorsense_pattern = re.compile(r'tremorsense +([0-9]+)')
truesight_pattern = re.compile(r'truesight +([0-9]+)')

df['passive perception'] = df['sense'].map(find(pass_perception_pattern))
df['blindsight'] = df['sense'].map(find(blindsight_pattern))
df['darkvision'] = df['sense'].map(find(darkvision_pattern))
df['tremorsense'] = df['sense'].map(find(tremorsense_pattern))
df['truesight'] = df['sense'].map(find(truesight_pattern))

# in order to generalize the speed, we summed up the speed for each row and replaced the specific type of speed with a boolean demonstrating what speed was used

# first change none to 0
def none_to_zero(x):
  if 'none' == x:
    return 0
  return x

df['speed'] = df['speed'].map(none_to_zero)
df['burrow speed'] = df['burrow speed'].map(none_to_zero)
df['climb speed'] = df['climb speed'].map(none_to_zero)
df['flying speed'] = df['flying speed'].map(none_to_zero)
df['swim speed'] = df['swim speed'].map(none_to_zero)

# sum and change to boolean
def add(lst):
  total = 0
  for n in lst:
    total += int(n)
  return total

def above_zero_boolean(feature):
  def convert(x):
    if int(x[feature]) > 0:
      return 1
    return 0
  
  return convert

df['total speed'] = df.apply(lambda x: add([x['speed'], x['burrow speed'], x['climb speed'], x['flying speed'], x['swim speed']]), axis=1)
df['has walk'] = df.apply(above_zero_boolean('speed'), axis=1)
df['has burrow'] = df.apply(above_zero_boolean('burrow speed'), axis=1)
df['has climb'] = df.apply(above_zero_boolean('climb speed'), axis=1)
df['has flying'] = df.apply(above_zero_boolean('flying speed'), axis=1)
df['has swim']  = df.apply(above_zero_boolean('swim speed'), axis=1)

# count the number of statuses a monster has been given for each category
def count_status(x):
    if x == 'none':
      return 0
    total_lst = []
    count_lst = x.replace(';', ',').replace('.', ',').replace('and', ',').split(',')
    for w in count_lst:
      w = w.strip().lower()
      if w in string.punctuation or w == '':
        continue
      total_lst.append(w)
    
    return len(total_lst)

df['damage vulnerability'] = df['damage vulnerability'].map(count_status)
df['damage resistance'] = df['damage resistance'].map(count_status)
df['damage immunity'] = df['damage immunity'].map(count_status)
df['condition immunity'] = df['condition immunity'].map(count_status)

# make num legendary actions and legendary resistance to 0 if none
df['num legendary actions'] = df['num legendary actions'].map(none_to_zero)
df['legendary resistance'] = df['legendary resistance'].map(none_to_zero)
# then split the legendary actions into actions that cost one, two, and three legendary actions
def count_matched(pattern):
  def count(x):
    lst = re.findall(pattern, x.lower())
    return len(lst)
  return count

def count_non_match(patterns):
  def count(x):
    lst = ast.literal_eval(x)
    c = 0
    for l in lst:
      if isinstance(l, str):
        not_matched = True
        for p in patterns:
          if re.search(p, l.lower()):
            not_matched = False
            break
        if not_matched:
          c += 1
      elif isinstance(l, tuple):
        not_matched = True
        for p in patterns:
          if re.search(p, l[1].lower()):
            not_matched = False
            break
        if not_matched:
          c += 1
    return c
  return count

df['legendary action cost 1'] = df['legendary actions'].map(count_non_match(['(?:costs? +)?(?:2|two)[A-Za-z ]+actions?', '(?:costs? +)?(?:3|three)[A-Za-z ]+actions?']))
df['legendary action cost 2'] = df['legendary actions'].map(count_matched('(?:costs? +)?(?:2|two)[A-Za-z ]+actions?'))
df['legendary action cost 3'] = df['legendary actions'].map(count_matched('(?:costs? +)?(?:3|three)[A-Za-z ]+actions?'))

# change reactions to count number of reactions
def count_actions(x):
  lst = ast.literal_eval(x)
  return len(lst)

df['reactions'] = df['reactions'].map(count_actions)

# count number of damaging actions
def count_attacks(x):
  lst = ast.literal_eval(x)
  c = 0
  for l in lst:
    if isinstance(l, tuple):
      c += 1
  return c

df['attack actions'] = df['actions'].map(count_attacks)

# find average damage monster can do with attacks
def average_dmg(x):
  lst = ast.literal_eval(x)
  c = 0
  dmg = 0
  for l in lst:
    if isinstance(l, tuple):
      c += 1
      damages = ast.literal_eval(l[7])
      temp = sum([int(d[0]) for d in damages])
      if l[8] == 'and':
        temp /= len(damages)
      dmg += temp

  return dmg/c if c > 0 else 0

df['average attack damage'] = df['actions'].map(average_dmg)

# change actions to counting number of actions the monster has
df['actions'] = df['actions'].map(count_actions)

# remove unneeded columns for columns which values were preprocessed to a different column
unneeded_columns = ['saving throws proficiencies', 'skill proficiencies', 'sense', 'speed', 'burrow speed', 'climb speed', 'flying speed', 'swim speed', 'legendary actions']
df = df.drop(columns=unneeded_columns)

df
```

## Exploring the Data

In order to better understand our data, we wanted to explore how the different stats retrieved from the website correlated with the challenge rating. By doing so, we hope to find general trends in how people decide stats for monsters.

#### Challenge Rating Distribution

Before doing any analysis, we decided to first look at the distribution of our data. As can be seen, our data seems to be skewed toward monsters with lower challenge ratings. This might be due to the fact that higher challenge rating monsters are treated as bosses which appear infrequently throughout a campaign. Therefore a lower number of these monsters were created. Another theory as to why a low number of high challenge rating monsters occurred could also be due to the fact that shorter campaigns occur more frequently, leading to high challenge rating monsters to be rarely used as players never level up that high. Either way, values that are derived from high challenge ratings are likely to be more varied.

```
import matplotlib.pyplot as plt

def swap(a, b, list_):
  tem = list_[a]
  list_[a] = list_[b]
  list_[b] = tem
  return list_

#clean up 
cr = df['challenge rating'].to_list()
size = df['size'].to_list()
cr_labels = list(set(cr))
cr_labels.sort()
cr_labels = swap(1,4,cr_labels)
cr_labels = swap(2,3,cr_labels)
first_four = cr_labels[0:4]
cr_labels = [int(a) for a in cr_labels[4:]]
cr_labels.sort()


for c in range(3, -1, -1):
  cr_labels.insert(0,first_four[c])
cr_labels = [str(a) for a in cr_labels]
size_labels = list(size_map.keys())
cr_size_distribution = {c:{s:0 for s in [1,2,3,4,5,6]} for c in cr_labels}
for c in range(len(cr)):
  s = size[c]
  c_ = cr[c]
  cr_size_distribution[c_][s] += 1

total_monster_per_cr = []
for c in cr_size_distribution:
  sum_ = sum(cr_size_distribution[c].values())
  total_monster_per_cr.append(sum_)
  for s in cr_size_distribution[c]:
    cr_size_distribution[c][s] = cr_size_distribution[c][s]/sum_ * 100

def getVal(dist, s):
  results = []
  for c in dist:
    results.append(dist[c][s])
  return results


plt.figure(figsize=(18,10))
plt.bar(range(len(cr_labels)),total_monster_per_cr,width=.5)
plt.xlabel('Challenge Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Challenge Ratings', fontweight="bold")
plt.xticks(range(len(cr_labels)), cr_labels)
plt.show()
```

#### Size vs Challenge Rating

Looking at the size proportion across the different challenge ratings, we can see that higher challenge ratings had larger sizes. This is likely because larger monsters are more intimidating and thus more suitable for fights against an experienced party. 

Even though this is the case, medium sized monsters can still generally be found across the the different challenge ratings and large sized monsters can be found for challenge ratings above 0. The only outliers to both of these observations is 22 and 27 which is comprised mostly of huge and gargantuan monsters. Additionally, challenge rating 29 doesn't contain any medium sized monsters. These anomalies could be due to the small distribution of monsters for these challenge ratings (as seen in the previous graph). 

```
plt.figure(figsize=(18,10))

bars = []
i = 0
last_s = None
for s in range(1,7):
  bar_ = np.array(getVal(cr_size_distribution, s))
  bars.append(plt.bar(range(len(cr_labels)),bar_,width=.5,bottom=last_s))
  if last_s is not None:
    last_s += bar_
  else:
    last_s = bar_

plt.xlabel('Challenge Rating')
plt.ylabel('Frequency (%)')
plt.title('Size Distribution of Challenge Ratings', fontweight="bold")
plt.xticks(range(len(cr_labels)), cr_labels)
plt.yticks(np.arange(0, 140, 10))
plt.legend([b[0] for b in bars], size_labels)
plt.show()
```

#### Ability Scores vs Challenge Rating

As to be expacted, the ability scores of a monster (strength, dexterity, constitution, intelligence, wisdom, and charisma) increase as the challenge rating increases. Although, there seems to be a greater emphasis on the strength and constitution scores. This is likely because strength helps increase the attack damage done by the monster and constitution helps increase the monster's hp. These factors are both vital when it comes to having the monster create a challenging fight for a party without worrying about the monster dying off too early.

## Model Hyperparameter Tuning, Training, and Testing

Because the classes are the challenge ratings which are initially read in as strings, we decided to go with Random Forest Classifier. The challenge rating (CR) themselves are whole numbers with specific meanings and implications. Thus, it is treated categorically and CR label is changed to ordinal values.

One important thing discovered in our review of our data set was how imbalanced it was. The monsters of high CR were outnumbered largely by the monsters of lower CRs. This will cause the high CR monsters to be less prioritized by the model, thus performing poorly on this minority class. To bypass this issue, we used the Synthetic Minority Oversampling Technique (SMOTE). It's a method of handling imbalance in datasets by oversampling the minority class. It takes the k-nearest neighbors of a data point within the minority class and linearly interpolates new values for that class. By adding more values into the minority class, it can be balanced with that majority class.

Below is our code implementation of the model hyperparameter tuning, training and testing as well as the use of SMOTE to handle our imbalanced dataset.

```
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from joblib import dump
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

feats = df.drop(columns=["name","challenge rating"])
labels = df['challenge rating']
label_map = {cr_labels[i]:i for i in range(len(cr_labels))}
labels = np.array([label_map[i] for i in labels])

oversample = SMOTE(k_neighbors=5)
feats,labels = oversample.fit_resample(feats.to_numpy(),labels)

print("Hyperparameter Tuning: n_estimators")
skf = StratifiedKFold(n_splits=5)
n_estimators = [100,250,500,1000,2000]
i = 0
accuracy_dict = {}
for train_index, test_index in skf.split(feats, labels):
  n = n_estimators[i]
  feats_train, feats_test = feats[train_index], feats[test_index]
  labels_train, labels_test = labels[train_index], labels[test_index]
  classifier = RandomForestClassifier(n_estimators=n)
  classifier.fit(feats_train, labels_train)
  pred = classifier.predict(feats_test)
  accuracy_dict[n] = accuracy_score(labels_test, pred)
  i += 1

print(accuracy_dict)

n = max(accuracy_dict, key=accuracy_dict.get)
print("----")
print("Model Training and Testing")
feats_train, feats_test, labels_train, labels_test = train_test_split(feats, labels,test_size=.2)
model = RandomForestClassifier(n_estimators=n)
model.fit(feats_train, labels_train)
pred = model.predict(feats_test)
print("Accuracy:",accuracy_score(labels_test, pred))

#Saving Model to a file
filename = 'model.joblib'
dump(model, filename)
```

### Other Model Attempts

The challenge rating (CR) themselves are in fact numbers so it could be called to question whether a regression algorithm would have been better suited for this task however. However, the numbers used to specify CR are mostly made up of whole numbers and fractions so it's not continuous, thus a regression algorithm would not have been ideal. As well, the numbers aren't exactly evenly spaced. Starting at 0, the CR increases to 1/8 then 1/4 then 1/2 before reaching whole numbers.

Below is cross validations with multiple regression models (Linear, Logistic, Ridge, and Lasso). Because the predicted values would be continuous, a plot of residuals was made for each graph.

```
from sklearn.linear_model import Ridge, Lasso, LogisticRegression

skf = StratifiedKFold(n_splits = 5)

accuracy_dict_other = {"linear":{},"logistic":{},"ridge":{},"lasso":{}}
i = 0

lambda_and_c_vals = [10 ** i for i in range(-3, 2)]

for train_index, test_index in skf.split(feats, labels):
  l_c = lambda_and_c_vals[i]
  feats_train, feats_test = feats[train_index], feats[test_index]
  labels_train, labels_test = labels[train_index], labels[test_index]
  
  linear = LinearRegression()
  linear.fit(feats_train, labels_train)
  pred = linear.predict(feats_test)
  resid = labels_test - pred
  plt.scatter(np.arange(1, len(resid) + 1), resid)
  print("Fold", i+1)
  plt.title("Linear Regression")
  plt.xlabel("i")
  plt.ylabel("Residuals")
  plt.show()

  logistic = LogisticRegression(C=l_c, max_iter=1e8,multi_class='ovr')
  logistic.fit(feats_train, labels_train)
  pred = logistic.predict(feats_test)
  resid = labels_test - pred
  plt.scatter(np.arange(1, len(resid) + 1), resid)
  plt.title("Logistic Regression: C = %f" % l_c)
  plt.xlabel("i")
  plt.ylabel("Residuals")
  plt.show()

  ridge = Ridge(alpha=l_c, max_iter=1e8)
  ridge.fit(feats_train, labels_train)
  pred = ridge.predict(feats_test)
  resid = labels_test - pred
  plt.scatter(np.arange(1, len(resid) + 1), resid)
  plt.title("Ridge Regression: Lambda = %f" % l_c)
  plt.xlabel("i")
  plt.ylabel("Residuals")
  plt.show()

  lasso = Lasso(alpha=l_c, max_iter=1e8)
  lasso.fit(feats_train, labels_train)
  pred = lasso.predict(feats_test)
  resid = labels_test - pred
  plt.scatter(np.arange(1, len(resid) + 1), resid)
  plt.title("Lasso Regression: Lambda = %f" % l_c)
  plt.xlabel("i")
  plt.ylabel("Residuals")
  plt.show()

  i += 1
```

As shown above the graphs themselves show no general trend or correlation. However, a lot of these graphs have obvious outliers in the data as their residuals are so large compared to the rest. For the most part though, the residuals lie within the +/- 5 range. However, 5 is a relatively large margin of error. So, it couldn't just be assumed that the predictions were "close enough."

### Final Thoughts

Overall, the data has shown that challenge rating isn't determined by one item on the stat block alone. As said earlier, all features were used because they all play their own role in strengthening a monster. There was never one feature that could summarize or characterize a monster's entire strength. It makes sense that way because there are so many factors of strength and durability Dungeons and Dragons has defined. For example, a flameskull is of CR 4 and is considered tiny with an average damage of 10. Though that size and damage may not seem intimidating, it’s considerably powerful. It has 11 resistances and immunities. As well, it has flying speed so it can attack from above which is advantageous. Despite its size and low damage, it makes up in other areas enough to be a strong contender for Level 4 players.

So when choosing a monster it's important to look over the stat block in detail. As a Dungeon Master, you have control over what your players can or can't do. First and foremost, ensure your CR is on level with your party. You don't want your monster to be overly powerful that it threatens a total party knock out. Secondly, ensure that the monster is durable against your players. Using the CR is a good guide, but having a strong monster that can deal appropriate damage isn't everything. That monster also needs to last for a decent amount of time against your players (as shown by what stats other DMs have emphasized). Lastly, take your party's strength into account. At the end of the day, the CR generated by our model can act as a stepping stone to gauge how strong a monster is. Now it's up to you to deteremine whether the monster is suitable for the party you lead. 

With that, happy adventuring on your DnD campaign!
